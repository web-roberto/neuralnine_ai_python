```python
import asyncio
import random

async def print_names():
    for name in ("John", "Kate", "Mike", "Alex", "Ann"):
        print(name)
        await asyncio.sleep(random.uniform(0.5, 1.5))# el sleep asíncrono es este time.sleep() es síncrono

async def print_ages(min_sleep, max_sleep):
    for _ in range(5):
        print(random.randint(20, 50))
        await asyncio.sleep(random.uniform(min_sleep, max_sleep))

loop = asyncio.get_event_loop() #creo un 'event loop' para ejecutar el asyncio
tasks = [print_names(), print_ages(0.2, 1)]# lista de tareas asíncronas a unir en un .gather(), aunque use () no está  llamando a la función pq son asícronas; si fueran síncronas, el () las llamaría
print tastks → corroutine object y corroutine object (no sé pq pero es así)
#Las corrutinas son las tareas que se pasan al event loop, se llamas así pq son colaborativas(asíncronas) que se lanzan en el event loop, se suspenden y devuelven el control al event loop(ver videos de Corey al final)
loop.run_until_complete(asyncio.gather(*tasks))# sobre el'event loop' ejecuto un .gather(coleccion) de tareas asíncronas (async/await)
# el *tasks es un unpacking o desestructuración de una lista
loop.close()#destruye el 'event loop'
```
